```python
r = np.linspace(5,9,200)
plt.plot(r, f.lennard_jones(r, 8, shifted=False) / (pc.joules_to_u√Öfs * pc.e) * 1000, label="Potentiel non-tronqu√©")
r = np.linspace(5,8,200)
plt.plot(r, f.lennard_jones(r, 8, shifted=True) / (pc.joules_to_u√Öfs * pc.e) * 1000, label="Potentiel tronqu√©")
plt.ylim(top=1, bottom=-2)
plt.xlim(5,9)
plt.ylabel("U (meV)")
plt.xlabel("r (√Ö)")
plt.axhline(y=0, linestyle="--", color="k")
plt.axvline(8, linestyle="dotted", color="r")
plt.text(x=8.2, y=-0.4, s=r"$r_c$", fontsize=15)
plt.legend()
plt.savefig("LJ_zoom.png", dpi=300)
```
