```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

1) $t \sim (O^N)$

```python
# Ewald, NVE
t = np.load("..\\Output\\Temps_Ewald_NVE_a30_rc10\\t_vs_N.npy")
N = np.load("..\\Output\\Temps_Ewald_NVE_a30_rc10\\N.npy") 
plt.scatter(N, t)
plt.yscale("log")
plt.xscale("log")
p, log_a = linregress(np.log10(N), np.log10(t))[:2]
n = np.linspace(0, np.max(N) + 500, 300)
plt.plot(n, 10**log_a * n**p, color="k", linestyle="--", linewidth=1, label=f"$t = {10**log_a:.3f}N^{{{p:.3f}}}$")
plt.legend()
plt.xlim(0, 500)
```

```python
# Coulomb, NVE
t = np.load("..\\Output\\Temps_Coulomb_NVE_a30_rc15\\t_vs_N.npy")
N = np.load("..\\Output\\Temps_Coulomb_NVE_a30_rc15\\N.npy") 
plt.scatter(N, t)
plt.yscale("log")
plt.xscale("log")
p, log_a = linregress(np.log10(N), np.log10(t))[:2]
n = np.linspace(0, np.max(N) + 500, 300)
plt.plot(n, 10**log_a * n**p, color="k", linestyle="--", linewidth=1, label=f"$t = {10**log_a:.3f}N^{{{p:.3f}}}$")
plt.legend()
plt.xlim(0, 500)
```

```python
# Coulomb, NVE
fig, axes = plt.subplots(sharey=True, figsize=(7,5))
t = np.load("..\\Output\\Temps_Coulomb_NVE_a30_rc15_no_parallel\\t_vs_N.npy")
N = np.load("..\\Output\\Temps_Coulomb_NVE_a30_rc15_no_parallel\\N.npy") 
axes.scatter(N, t, marker="s", label="Coulomb")
axes.set_yscale("log")
axes.set_xscale("log")
p, log_a = linregress(np.log10(N), np.log10(t))[:2]
n = np.linspace(0, np.max(N) + 500, 300)
axes.plot(n, 10**log_a * n**p, color="k", linestyle="--", linewidth=1, label=f"$t = {10**log_a:.3f}N^{{{p:.3f}}}$")
axes.legend()
axes.set_xlabel(r"$N$")
axes.set_ylabel(r"$t$ (s)")
axes.set_xlim(0, 500)

# Ewald, NVE
t = np.load("..\\Output\\Temps_Ewald_NVE_a30_rc15_no_parallel\\t_vs_N.npy")
N = np.load("..\\Output\\Temps_Ewald_NVE_a30_rc15_no_parallel\\N.npy") 
axes.scatter(N, t, label="Ewald")
axes.set_yscale("log")
axes.set_xscale("log")
p, log_a = linregress(np.log10(N), np.log10(t))[:2]
n = np.linspace(0, np.max(N) + 500, 300)
axes.plot(n, 10**log_a * n**p, color="k", linestyle="dashdot", linewidth=1, label=f"$t = {10**log_a:.3f}N^{{{p:.3f}}}$")
axes.legend()
axes.set_xlim(0, 500)
plt.savefig("Ordre_temporel.png", dpi=300)
```
