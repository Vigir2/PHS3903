```python
import numpy as np
import matplotlib.pyplot as plt
import forces as f
import parameters.h2O_model as h2O
import parameters.physical_constants as pc
```

```python
r = np.linspace(3,10,200)
plt.plot(r, f.lennard_jones(r, 10, shifted=False) / (pc.joules_to_uÅfs * pc.e) * 1000)
plt.ylim(top=10, bottom=-10)
plt.ylabel("U (meV)")
plt.xlabel("r (Å)")
plt.axhline(y=0, linestyle="--", color="k")
plt.axvline(8, linestyle="dotted", color="r")
plt.text(x=8.2, y=-1, s=r"$r_c$", fontsize=15)
plt.savefig("LJ.png", dpi=300)
```

```python
r = np.linspace(5,9,200)
plt.plot(r, f.lennard_jones(r, 8, shifted=False) / (pc.joules_to_uÅfs * pc.e) * 1000, label="Potentiel non-tronqué")
r = np.linspace(5,8,200)
plt.plot(r, f.lennard_jones(r, 8, shifted=True) / (pc.joules_to_uÅfs * pc.e) * 1000, label="Potentiel tronqué")
plt.ylim(top=1, bottom=-2)
plt.xlim(5,9)
plt.ylabel("U (meV)")
plt.xlabel("r (Å)")
plt.axhline(y=0, linestyle="--", color="k")
plt.axvline(8, linestyle="dotted", color="r")
plt.text(x=8.2, y=-0.4, s=r"$r_c$", fontsize=15)
plt.legend()
plt.savefig("LJ_zoom.png", dpi=300)
```

```python
r = np.linspace(0.1,14,200)
plt.plot(r, f.coulomb(r, 8, h2O.q, h2O.q)/ (pc.joules_to_uÅfs * pc.e))
r = np.linspace(0.1,8,200)
plt.plot(r, f.coulomb(r, 8, h2O.q, h2O.q, shifted=True)/ (pc.joules_to_uÅfs * pc.e))
plt.xlim(0,14)
plt.ylim(top=10, bottom=-1)
plt.ylabel("U (eV)")
plt.xlabel("r (Å)")
plt.axvline(8, linestyle="dotted", color="r")
plt.text(x=8.2, y=-0.6, s=r"$r_c$", fontsize=15)
plt.axhline(y=0, linestyle="--", color="k")
plt.savefig("Présentation/Coulomb.png", dpi=300)
```
